```python
df['tags'] = df['overview']+ " "  + df['genres']+ " " + df['tagline']
```

## **NLP Text preprocessing**

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re                                   # re means Regular expressions
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  # lowercase
  text = str(text).lower()
  # remove the punctuations
  text = re.sub(r'[^\w\s]', '', text)

  words = text.split()
  # remove stopwords
  words = [word for word in words if word not in stop_words]

  # lemmatize
  words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]

  return ' '.join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df['tags'][1]
```

## **Text Vectorization**

```python
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from nltk.translate.ribes_score import ngrams
tfidf = TfidfVectorizer(max_features=50000, ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

## **Testing and making the pickle file**


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 10):
  if title not in indices:
    return ['Movies not found']

  idx = indices[title]
  sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('Toy Story')
```

```python
import pickle

pickle.dump(tfidf_matrix,open('tfidf_matrix.pkl','wb'))

pickle.dump(indices,open('indices.pkl','wb'))

df.to_pickle('df.pkl')

pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
